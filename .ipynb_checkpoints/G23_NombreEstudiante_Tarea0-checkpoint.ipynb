# Objetivo
Conocer e instalar un entorno para programar y documentar en ciencia de datos.

# Ejercicio 1

Instalación de las herramientas de programación.

Se descarga e instala Pyhton3 desde https://www.python.org/downloads/, después se descarga e instala la distribución de Anaconda desde https://www.anaconda.com/download. Para reconecer los paquetes se ejecuta en Anaconda prompt el comando `conda list` y se verifica la instalacción de :

- NumPy
- Pandas
- Seaborn
- matplotlib
- sklearn
- requests
- Bokeh*
- Pip
- folium

Después se actualiza Pandas con `pip install --upgrade pandas`

Se abre *jupyter notebook* desde la interfaz de Anaconda, se crea una carpeta y el primer notebook para la tarea 0.

*Notas:* 

- Comprobación de versión de Anaconda: ` anaconda --version`



- Comprobación de paquetes directamente en el notebook:
  
    `%load_ext version_information`
  
    `%version_information numpy,pandas,matplotlib, seaborn, sklearn, request, bokeh , pip, folium`
