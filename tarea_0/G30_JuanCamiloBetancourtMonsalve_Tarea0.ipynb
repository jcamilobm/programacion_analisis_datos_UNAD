## 3.1
Construya un diccionario en el cual las llaves (keys) sean los
nombres (strings) de ocho (8) personas de su entorno familiar o
amigos y los valores (values) sean tuplas de cuatro (4)
elementos que contengan la siguiente información de cada
persona en orden: (Edad, Peso, Estatura, Estudia). La edad debe
ser en años, el peso en kilogramos, la estatura en metros y el
elemento de estudia debe ser un Booleano, en el cual True
significa que si estudia y False el caso contrario.

```python
dict = {
     'luis felipe' : (34, 80 , 1.75, False) ,
     'geancarlo' : (12, 42 , 1.65, True) ,
     'gloria' : (65, 62 , 1.57, False) ,
     'angel' : (72, 70 , 1.61, False) ,
     'sebastian' : (33, 55 , 1.75, True) ,
     'luciana' : (7, 40 , 1.40, True) ,
     'christian' : (26, 65 , 1.67, True) ,
     'lucas' : (16, 55 , 1.75, False) ,
}
```

## 3.2
Implemente ciclos (for o while) para extraer la información deldiccionario creado anteriormente y generar las siguientes cuatro
(4) listas: Edades, Pesos, Estaturas y Estudia.

La estructura base para recorrer un diccionario es:

`{key:value for (key,value) in dictonary.items() }`

```python
# Preparo los contenedores
Edades,Pesos,Estaturas,Estudia = [],[],[],[]

# Recorro el diccionario teniendo en cuenta el unpacking de tuplas. Nota: <name_dic>.items() -> [ (key1,(value1,value1)) , (keyn,(valuen,valuen)) ,...]
for key , (edad,peso,estatura,estudia) in dict.items():
    #agrego el valor respectivo en al última posición de cada contenedor con el método append:  <secuencia>.append(elemento)
    Edades.append(edad)
    Pesos.append(peso)
    Estaturas.append(estatura)
    Estudia.append(estudia)

# Compruebo solución
print(f'Edades: {Edades}')
print(f'Pesos: {Pesos}')
print(f'Estaturas: {Estaturas}')
print(f'Estudia: {Estudia}')
```

## 3.3


Defina una función que reciba como parámetros una lista y
regrese (return) el promedio de dicha lista. Utilice dicha función
para imprimir (print) el promedio de las listas del inciso anterior


```python
def hallar_promedio(my_list):
    """
    Esta función recibe una lista de números y devuelve su promedio.
    """
    sum , n = 0 , 0 # unpacking de tupla
    for value in my_list:
          sum += value
          n += 1
    return round(sum / n , 2 )
```

```python
# llamar función con lista de Edades e imprimirla
print( "Promedio Edudades: " , hallar_promedio(Edades))
print( "Promedio Pesos: " , hallar_promedio(Pesos))
print( "Promedio Estaturas: " , hallar_promedio(Estaturas))
print( "Promedio Estudia: " , hallar_promedio(Estudia))
```

```python
# Python ofrece funciones propias para facilitar los calculos:

def hallar_promedio_v2(my_list):
    return round( sum(my_list) / len(my_list) , 2 )
```

```python
# llamar función con lista de Edades e imprimirla
print( "Promedio Edudades: " , hallar_promedio(Edades))
print( "Promedio Pesos: " , hallar_promedio(Pesos))
print( "Promedio Estaturas: " , hallar_promedio(Estaturas))
print( "Promedio Estudia: " , hallar_promedio(Estudia))
```
