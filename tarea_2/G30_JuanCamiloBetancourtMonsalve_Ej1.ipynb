# Teoría




- **¿Qué es REST web service y cómo se utilizan?**

  **REST**(**Re**presentational **S**tate **T**ransfer) es un estilo de arquitectura de software para aprovechar las tecnologías y protocolos de la web. Es un conjunto de reglas de cómo crear una API WEB.

  Se basa en un único protoloco: HTTP, y se puede utilizar para acceder y manipular datos facilmente con verbos como get,post,put y delete.

  Se caracteriza por devolver los resultados en formatos ligeros como JSON y XML.


image.png

- **¿Qué es una RESTful API?**

  RESTful API son servicios web que implementan la arquitectura REST.
  Tiene principalmente tres componentes:

  1.   URL endpoint
  2.   HTTP verb
  3.   Body

**URL endpoint**: Es un enlace que representa el recurso que se requiere acceder y manipular. Esos recursos pueden ser texto, imagenes etc.

  Ejemplo endpoint: segurosad/afiliados

**HTTP verb** : Los verbos definen las acciones sobre los recursos.

    -  Get: Obtener recursos
    -  Post: Crear un recurso
    -  Delete: Borrar un recurso
    -  Put: Actualizar un recurso.


   **Body**: Al utilizar el verbo POST o Put en el body se pasa un JSON con datos necesarios de la API.

  **Header**: En el header se puede incluir la autorización de la API, el tipo de archivos que acepta el consumo,etc.



- **¿Qué es JSON?**

**JSON**( (JavaScript Object Notation - Notación de Objetos de JavaScript) ) es un formato ligero para intercambiar datos.
Es una cadena que puede incluir cadenas,números,arreglos,booleanos y otros literales.

# Consumo de API

Desarrollar un código de python que permita obtener los datos
de las naves de SpaceX, desde la API spacexdata v4:
https://api.spacexdata.com/v4/ships, considerando:


*  request.get
*  Response.json



````python
import requests

#URL de la API SapcesX
url = 'https://api.spacexdata.com/v4/ships'

# petición GET

response = requests.get(url)
````

````python
# validaciones
if response.status_code == 200:
  #datos
  data = response.json()

else:
  print(f"Error: {response.status_code}")

#data
````

# Cargar JSON como un DataFrame
Cargar el json de la respuesta del request anterior como un
DataFrame de pandas (ayuda: función pd.DataFrame.from_dict).


````python
df = pd.DataFrame.from_dict(data)
df.head()
````

# Cantidad de columnas de un DataFrame

Determinar cuantas columnas tiene el DataFrame anterior, los
tipos de datos (Dtype) que tiene y los datos ausentes.

````python
print('Número de columnas: ',df.shape[1])
````

````python
# tipos de datos y datos ausentes
df.info()
````

````python
# Calcular la cantidad de valores nulos en cada columna
# null_counts = df.isnull().sum().sort_values()
null_proportion = ( df.isnull().sum() /df.shape[0] ).sort_values()

# Crear el histograma
plt.figure(figsize=(10, 6))
null_proportion.plot(kind='bar', color='skyblue')
plt.title('Proporción de Valores Nulos por Columna (Ordenado)')
plt.xlabel('Columnas')
plt.ylabel('Proporción de Valores Nulos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

**Interpretación**: Según el histograma a partir del atributo abs se tienen muchos valores nulos para considerar esas variables para un análisis por lo que se descartan.
