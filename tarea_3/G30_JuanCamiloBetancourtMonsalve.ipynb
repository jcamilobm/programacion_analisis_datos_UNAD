# Introducción

El **análisis exploratorio de datos (AED)** es un enfoque para resumir, visualizar y familiarizarse con las características relevantes de un conjunto de datos. El objetivo principal  es ayudar a analizar los datos  antes de hacer suposiciones.

Algunos de los métodos que utiliza son:

*   Visualización univariada
*   Visualización Bivariada
*   Visualización multivaridad

**3. Proceso**

- Resumenes: Calcular media,mediana,moda,desviación estándar,etc.Para columnas categoricas calcular la frecuencia.

- Limpieza: Revisar valores faltantes y cómo manejarlos.

- Detectar valores atípicos mediante diagramas de caja, Z-Score,IQR.

**Transformación de datos**

**Normalización**

Trata cada columna en una misma escala para que no haya predominancia.

-  Min-Max: Manejar los datos en una sola escala, generalmente entre o y 1.Preserva la esteructura de los datos, si se tiene datos atipicos se mantiene.
- Estandarización: Transformar los datos par que tengan una media de 0 y una desviación estándar de 1. Es importante apra algoritmos basados en distancias o gradientes.

- Log: Ayuda cuando los datos están muy sesgados. Reduce el impacto de valores atípicos.

**Encoding**

Convertir tipos de datos categoricos en numericos.

-  One-Hot Encoding: Crea una nueva columna por cada categoría e indica la presencial con 1 o 0.
- Label Encoding: Transforma en uan secuencia de números. Importante si hay una relación ordinal en la categoría.

**Data Visualization**

- Histogramas: Visuaalizar distribución de variables numéricas.
- Heapmap: Visualizar cómo diferentes variables interactuan.
- Pairplots: Describe realciones por apres de una manera agil.

## Estandarización/ Escalamiento

Estandarice o escale al menos dos variables numéricas que sean diferentes en
magnitud o unidades de medición, de modo que pueda comparar y encontrar
similitudes o diferencias en sus distribuciones (visualmente, ver concepto 3).

**MinMaxScaler**(feature_range = (0, 1)): transformará cada valor en la columna proporcionalmente dentro del rango [0,1]. Mantiene la forma del conjunto de datos (sin distorsión). 🚩Sensitive to outliers

**StandardScaler**() transformará cada valor en la columna para oscilar alrededor de la media 0 y la desviación estándar 1. Se Utiliza StandardScaler si sabe que la distribución de datos es normal.

Si hay valores atípicos, se sugiere **RobustScaler**(). Se basa en percentiles y, por lo tanto, no están influenciadas por una pequeña cantidad de valores atípicos marginales muy grandes.

**Elección escalador para el Dataset**

El dataset contiene muchos valores atipípicos por lo que se escoge el esacalado  *RobustScaler()*

Además, se identifica que hay muchas modas por que se compara el peso y la potencia pero por combustible.

```python
# Definir la lista de columnas a graficar
columnas_a_graficar = ['peso', 'potencia']

# Definir la lista de clases a filtrar
clases_a_filtrar = df["combustible"].unique()  # Obtener todas las clases únicas

# Estandarizar las variables
#scaler = StandardScaler()
#scaler = MinMaxScaler()
scaler = RobustScaler()

# Graficar las distribuciones para cada clase
for clase_filtrada in clases_a_filtrar:
    # Filtrar el DataFrame por la clase definida
    df_filtrado = df[df["combustible"] == clase_filtrada]

    #df_filtrado = df[(df["combustible"].isin( clases_a_filtrar)) & (df["clase"] == 'CAMIONETA')]
    # Seleccionar las columnas numéricas relevantes
    columnas_numericas = df_filtrado[columnas_a_graficar]

    # Estandarizar las variables
    scaled_data = scaler.fit_transform(columnas_numericas)
    #scaled_data =   np.log(columnas_numericas)

    # Crear un DataFrame con las variables estandarizadas
    scaled_df = pd.DataFrame(scaled_data, columns=columnas_numericas.columns)

    # Graficar las distribuciones de las variables estandarizadas para la clase actual
    plt.figure(figsize=(6, 3))
    for columna in columnas_a_graficar:
        sns.kdeplot(data=scaled_df[columna], fill=True, label=f'{columna} estandarizado')

    plt.title(f'Distribuciones estandarizadas para la clase {clase_filtrada}', fontsize=12)
    plt.xlabel('Valor estandarizado', fontsize=7)
    plt.ylabel('Densidad', fontsize=7)
    plt.legend(title='', fontsize=7)
    plt.show()
```

**Forma de la distribución**: Se observan múltiples modas por lo que hay varios grupos de vehículos.

# **Concepto 3: Gráficos y visualización científica**

# Diagrama de barras

**Objetivo:**

*   Cuál es la clase de vehículo predominante en Colombia?
*   Cuál es el combustible predominante por clase?

**Utilidad:** Las gráficas de barras son útiles para visualizar la distribución de una variable categórica y comparar las frecuencias entre diferentes categorías.

```python
# Calcular la frecuencia de cada categoría de marca
frecuencia_por_categoria = df.groupby(['clase', 'combustible']).size().unstack(fill_value=0)

# Ordenar las columnas (categorías) de cada fila (marca) de mayor a menor frecuencia
frecuencia_por_categoria = frecuencia_por_categoria.apply(lambda x: x.sort_values(ascending=False), axis=1)

# Crear el diagrama de barras
plt.figure(figsize=(10, 6))
frecuencia_por_categoria.plot(kind='barh', stacked=True)
plt.title('Frecuencia de clase de vehículo categorizzado por combustible')
plt.xlabel('Clase')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.legend(title='Combustible')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

La clase 'camioneta' es la predominante con el combustible híbrido Gasolina+Eléctrico.

Sin embargo el tricimoto,la motocicleta, el cuadriciclo, el camión y el bus tienen un combustible predominante ELECTRICO.

## Diagrama de pie

**Objetivo:** Cuál es el tipo de combustible predominante en Colombia entre los eléctricos/híbridos?

**Utilidad:** Los diagramas de pie (o gráficos circulares) son útiles para visualizar la proporción de cada categoría dentro de un conjunto de datos en relación con el total. Es conveniente para variables con pocas categorias.

A continuación, se mustra el diagrama de Pie de la variable '**combustible**' considerando que solo tiene 3 categorias lo que hace fácil la interpretación de proporción.

```python
# Contar los valores de cada categoría
conteo_clasificacion = df['combustible'].value_counts()

# Crear el gráfico de pie
plt.figure(figsize=(10, 6))
plt.pie(conteo_clasificacion, labels=[''] * len(conteo_clasificacion), startangle=140)
plt.title('Diagrama de Pie: Combustible')
plt.axis('equal')  # Para asegurar que el círculo sea circular

# Crear la leyenda con los nombres de las categorías y un código de color
for idx, (categoria, conteo) in enumerate(conteo_clasificacion.items()):
    plt.text(1.4, 0.15 - idx * 0.1, f'{categoria}: {conteo}', fontsize=10, color='black')
    plt.gca().add_patch(plt.Circle((1.2, 0.15 - idx * 0.1), 0.03, color=f'C{idx}'))

plt.show()
```



*   Según el diagram de Pie la mayoría de los vehículos son híbridos (GASO ELEC)

Hallar longitud y latitud promedio de cada departamento de Colombia a partir de geometry.



*   **Latitude_DPT**
*   **Longitude_DPT**


**Utilidad:** Agregar marcadores en el centro de cada departamento en el mapa.


```python
latitudes = []
longitudes = []

for idx, row in gdf.iterrows():
    polygon = shape(row['geometry'])  # Convierte el polígono GeoJSON a un objeto Shapely
    centroid = polygon.centroid  # Calcula el centroide del polígono
    latitudes.append(centroid.y)  # Latitud del centroide
    longitudes.append(centroid.x)  # Longitud del centroide

# Agregar las coordenadas al DataFrame como nuevas columnas
gdf['Latitude_DPT'] = latitudes
gdf['Longitude_DPT'] = longitudes

gdf.head(3)
```

```python

# Se une el df corregido y el df de geo_json

join_df = pd.merge(df , gdf, left_on='departamento', right_on='NOMBRE_DPT', how='left')
join_df.head(3)
```

```python
gdf_count =  join_df.groupby(['NOMBRE_DPT','combustible']).agg({'NOMBRE_DPT':'count','Latitude_DPT':'first', 'Longitude_DPT':'first'}).rename(columns={'NOMBRE_DPT': 'count'}).reset_index()
#gdf_count =  join_df.groupby('NOMBRE_DPT').agg({'NOMBRE_DPT':'count','Latitude_DPT':'first', 'Longitude_DPT':'first'}).rename(columns={'NOMBRE_DPT': 'count'}).reset_index()
#gdf_count =  join_df.groupby('NOMBRE_DPT').agg({'NOMBRE_DPT':'count','combustible':'count', 'Latitude_DPT':'first', 'Longitude_DPT':'first'}).rename(columns={'NOMBRE_DPT': 'count'}).reset_index()
#gdf_count.head(3)

# Agrupar por 'NOMBRE_DPT' y 'combustible' y calcular la suma de 'count'
gdf_count = gdf_count.groupby(['NOMBRE_DPT', 'combustible'])['count'].sum().unstack(fill_value=0)

# Agregar una columna 'Total' que contiene la suma de los valores de cada fila
gdf_count['TOTAL'] = gdf_count.sum(axis=1)

# Restablecer el índice para hacer NOMBRE_DPT una columna nuevamente
gdf_count.reset_index(inplace=True)

# Combinar con los valores originales de Latitude_DPT y Longitude_DPT
gdf_count = gdf_count.merge(join_df[['NOMBRE_DPT', 'Latitude_DPT', 'Longitude_DPT']].drop_duplicates(), on='NOMBRE_DPT')
#gdf_count
```

 ### Visualización

```python
# Función para generar contenido HTML de la tabla en el tooltip
def generate_table_html(row):
    html = "<table style='width:100%; border-collapse: collapse;'>"
    html += "<tr style='background-color: #c7d2ff'>"
    html += "<th style='border: 1px solid black; padding: 8px;'>DEP</th>"
    html += "<th style='border: 1px solid black; padding: 8px;'>Total Veh</th>"
    html += "<th style='border: 1px solid black; padding: 8px;'>DIES ELEC</th>"
    html += "<th style='border: 1px solid black; padding: 8px;'>ELECTRICO</th>"
    html += "<th style='border: 1px solid black; padding: 8px;'>GASO ELEC</th>"
    html += "</tr>"
    html += f"<tr><td style='border: 1px solid black; padding: 8px;'>{row['NOMBRE_DPT']}</td><td style='border: 1px solid black; padding: 8px;'>{row['TOTAL']}</td><td style='border: 1px solid black; padding: 8px;'>{row['DIES ELEC']}</td><td style='border: 1px solid black; padding: 8px;'>{row['ELECTRICO']}</td><td style='border: 1px solid black; padding: 8px;'>{row['GASO ELEC']}</td></tr>"
    html += "</table>"
    return html
```
