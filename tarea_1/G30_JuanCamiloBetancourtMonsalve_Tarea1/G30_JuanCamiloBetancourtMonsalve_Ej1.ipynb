### Estructuras de datos nativas de Python

Según [Wes] :

**Tupla**

Las tuplas son secuencias *inmutables* de longitud fija.


```
my_tuple = (e,e2, ...,e_n)
my_tuple = (e,)
```

Pueden almacenar *elementos de diferente tipo*:


```
my_tuple = (2 , 2.5, True, "juan")
```

**Lista**

Secuencias con longitud variable y contenido modificable.

```
my_list = [e,e2, ...,e_n]

```

Como son mutables admiten:


```
my_list[n] = <new_value>

```


**Diccionario**

El diccionario almacena una colección de  clave:valor.
Cada clave está asociada con un valor para ser recuperado,insertado,modificado o eliminado.


```
my_dict = {key1:value1, ... , keyn:valuen}
```

Los valores del diccionario pueden ser cualquier objeto de Python, sin embargo las claves deben ser inmutables como los tipos escalares(int,float,string) o las tuplas y a su vez los elementos de la tupla deben ser inmutables.


```
my_dict2 = {  [1,3] : 'bad key' }  # Diccionario incorrecto.
```

**Set**

Un set es una colección sin orden de elementos únicos.

```
my_set = { value2 , valuen , ... , value1}
```

Soportan operaciones matemáticas como union,intersección ,diferencia y diferencia simétrica.

Algunas operaciones de los sets:



```
a = {1,2,3}
b = {1,5}

a.union(b)                  => {1,2,3,5}
a.intersection(b)           => {1}
a.difference(b)             => {2,3}
a.symmetric_difference(b)   => {2 , 3 , 5}

```

## Extraer encabezados de columnas de cada DataFrame del Panel


Programe una función que extraiga todas las listas de los
encabezados de columnas que se encuentran en un Panel y
úsela para extraer las listas de columnas del Panel del punto
anterior. Imprima los resultados de forma que se diferencien los
encabezados entre los dataframes en el Panel.

````python
def get_encabezados(panel):
  enc = {}
  for key, df in panel.items():
    enc[key] = df.columns
  return enc

def show(encabezados_panel):
  for key,value in encabezados_panel.items():
    print("==========  "+str(key)+" =============\n")
    for i in value:
      print("- "+ str(i))
    print("\n")
````

````python
# valido
enc = get_encabezados(panel)
show(enc)
````

````python
# Se puede organizar en un dataframe
def get_encabezados_v2(panel):
  dic = {}
  for key , value in panel.items():
    dic[key] =  pd.Series(value.columns)
  return pd.DataFrame(dic)
````

````python
# probar
df_encabezados = get_encabezados_v2(panel)
df_encabezados
````

## Cantidad de elementos de cada df en Panel


Programe una función que imprima la cantidad de elementos
que tienen las bases de datos en un Panel y úsela con el Panel
del punto anterior. Imprima los resultados de forma que se
diferencien los tamaños de los diferentes dataframes en el
Panel.

````python
def get_dimension(panel):
  dim = {}
  for key, df in panel.items():
    dim[key] = df.shape
  return dim
````

````python
dim = get_dimension(panel)
show(dim)
````
