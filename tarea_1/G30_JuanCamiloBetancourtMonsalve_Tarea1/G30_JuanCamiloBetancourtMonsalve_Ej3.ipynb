## Gr√°fico 1:


Mostrar los puntos de dispersi√≥n del emparejamiento las
siguientes listas:

$$x= [13,14,15,16,17,18]$$
$$y= [16,24,45,57,75,102]$$

‚Ä¢ Mostrar la funci√≥n lineal ùíá(ùíô) = ùüèùüïùíô ‚àí ùüêùüèùüé. ùüëùüëùüëùüë en el dominio

‚Ä¢ Mostrar la funci√≥n exponencial ùíà(ùíô) = ùüé. ùüèùüíùüìùüê ùíÜ(ùüé.ùüëùüîùüóùíô) en el
dominio [13,18].

‚Ä¢ Mostrar la funci√≥n logar√≠tmica ùíâ(ùíô) = ùüêùüìùüó. ùüíùüó ùíçùíè(ùíô) ‚àí ùüîùüìùüî. ùüíùüî en el dominio [13,18].

‚Ä¢ Los objetos mostrados deben tener colores diferentes.

‚Ä¢ La l√≠nea de la funci√≥n logar√≠tmica debe ser punteada.

‚Ä¢ En los ejes deben visualizarse los n√∫meros.

‚Ä¢ Asigne un nombre (label) a cada objeto e imprima la legenda
que los identifique en la gr√°fica.

‚Ä¢ Asigne nombres a los ejes de la gr√°fica.

‚Ä¢ Asigne un t√≠tulo a la gr√°fica.

‚Ä¢ Tenga en cuenta que los nombres que decida utilizar le ser√°n de
ayuda para presentar las conclusiones que usted obtenga del
ejercicio.

```python
# Defino puntos de dispersi√≥n
x = [13, 14, 15, 16, 17, 18]
y = [16, 24, 45, 57, 75, 102]

#Defino funciones
def f(x):
    return 17 * x - 210.3333

def g(x):
    return 0.1452 * np.exp(0.369 * x)

def h(x):
    return 259.49 * np.log(x) - 656.46

# Crear un conjunto de valores x en el dominio [13, 18] partidos en 100
x_v = np.linspace(13, 18, 100)

# configuro la figura
plt.figure(figsize=(10,6))
plt.xlabel("x")
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Puntos de dispersi√≥n')

# Trazo los puntos de dispersi√≥n
plt.scatter(x,y,label='Datos de dispersi√≥n', color='red')

plt.plot(x_v, f(x_v), label='$f(x) := 17x - 210.3333$',color='pink')

plt.plot(x_v, g(x_v), label='$g(x) := 0.1452 * e^{0.369x}$',color="green")

plt.plot(x_v, h(x_v), label='$h(x):= 259.49 * log(x) - 656.46$' , linestyle="--",color='blue')


# Muestro la leyenda
plt.legend()

# Muestro el gr√°fico
plt.grid(True)
plt.show()
```

**An√°lisis del gr√°fico 1**

- Python permite analizar graficas en una sola figura para sacar comparaciones valiosas.
- Las graficas son las regresiones de los puntos.

- La mejor regresi√≥n es la lineal seg√∫n el error cuadr√°tico medio.

```python
# Calcular el error cuadr√°tico medio (ECM)

def ecm(y_observado, y_prediccion):
    return  np.mean((y_observado - y_prediccion)**2)

ecm_f = ecm( np.array(y), np.array([f(x) for x in x]) )
ecm_g = ecm( np.array(y), np.array([g(x) for x in x]) )
ecm_h = ecm( np.array(y), np.array([h(x) for x in x]) )

print("Error cuadr√°tico medio f(x): ", ecm_f  )
print("Error cuadr√°tico medio g(x): ", ecm_g  )
print("Error cuadr√°tico medio h(x): ", ecm_h  )
```

El Error cuadr√°tico medio indica que la funci√≥n lineal f(x) predice mejor los resultados.

# Gr√°fico 2:
Construya los c√≥digos necesarios para:

- Generar 4 listas aleatorias de 700 muestras (revise la
documentaci√≥n de numpy.random):

  - x1 con distribuci√≥n normal de media 5 y desviaci√≥n
  est√°ndar 0.1.

  - y1 con distribuci√≥n normal de media 7 y desviaci√≥n est√°ndar 0.1.
 -  x2 con distribuci√≥n uniforme entre el valor m√≠nimo de x1 y
el valor m√°ximo de x1.

  - y2 con distribuci√≥n uniforme entre el valor m√≠nimo de y1 y
el valor m√°ximo de y1.

- Graficar en un mismo plot nubes de dispersi√≥n de los pares de
puntos que se forman con (x1, y1) y (x2, y2).

- La nube de puntos (x2, y2) debe visualizarse sobre la nube de
(x1, y1).

- Anexe los t√≠tulos y legendas necesarios para poder citar y
explicar realizar el an√°lisis del gr√°fico obtenido y las
conclusiones.

```python
# Definir los par√°metros de media y desviaci√≥n est√°ndar
media_x1 = 5
desviacion_estandar_x1 = 0.1

media_y1 = 7
desviacion_estandar_y1 = 0.1

# Generar las listas aleatorias
# loc: media o valor esperado
x1 = np.random.normal(loc=media_x1, scale=desviacion_estandar_x1, size=700)
y1 = np.random.normal(loc=media_y1, scale=desviacion_estandar_y1, size=700)

# Imprimir
print("x1:", x1[:5])
print("")
print("y1:", y1[:5])

# Hallar valores m√≠nimos

min_x1 ,max_x1 = x1.min() , x1.max()
min_y1 , max_y1 = y1.min() , y1.max()

print(f'Rango de x1: {min_x1} a {max_x1}')
print(f'Rango de  y1: {min_y1} a {max_y1}')
```

```python
# Generar un array con distribuci√≥n uniforme
x2 =  np.random.uniform(low=min_x1, high=max_x1, size=700)
y2 =  np.random.uniform(low=min_y1, high=max_y1, size=700)
# Imprimir
print("x2:", x2[:5])
print("")
print("y2:", y2[:5])
```

- Graficar en un mismo plot nubes de dispersi√≥n de los pares de
puntos que se forman con (x1, y1) y (x2, y2).

```python
# Graficar las nubes de dispersi√≥n
plt.scatter(x1, y1, color='blue',alpha=0.4,  label='Dist.Normal')
plt.scatter(x2, y2, color='red', alpha=0.6, label='Dist.Uniforme')

# Etiquetas y t√≠tulo del gr√°fico
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Nubes de Dispersi√≥n')

# Mostrar la leyenda
plt.legend()

# Mostrar el gr√°fico
plt.show()
```

**An√°lisis del gr√°fico 2**

- La distribuci√≥n normal aglomera los puntos alrededor de la media , por otro lado la uniforme los distribuye en todo el espacio.
- Python permite hacer simulaciones para notar patrones en los datos.
